2. [Deret MacLaurin](https://en.wikipedia.org/wiki/Taylor_series#Trigonometric_functions) dari $\arctan(x)$ adalah:
    $$
    \arctan(x) = \sum_{n = 0}^{\infty} \frac{(-1)^nx^{2n + 1}}{2n+1}
    $$
    Substitusi $x = 1$ memberikan representation deret dari nilai bilangan $\pi/4$. Buat program Python ringkas yang menghitung jumlah parsial hingga $N=5000$ untuk mengaproksimasi: 
    $$
    \frac{\pi}{4} \approx \sum_{n = 0}^{5000} \frac{(-1)^nx^{2n + 1}}{2n+1}
    $$

3. Buat suatu fungsi yang dinamakan `arctan_taylor` yang mengambil parameter masukan (*input*) `x` dan `N`. Fungsi ini kemudian memberikan keluaran (*output*) polinomial Taylor derajat $N$ dari fungsi $\arctan x$ yang dievaluasi pada `x`:
    $$
    \sum_{k=0}^N (-1)^k \frac{x^{2k + 1}}{2k + 1}
    $$
    Selanjutnya, ujilah fungsi tersebut untuk suatu nilai $x = 1$ dan $N = 5000$. **Catatan uji kewarasan ("sanity check")**: hasilnya harus konsisten dengan soal no. 2.

4. Buatlah suatu fungsi bernama `sqrt_integral` yang mengambil parameter masukan `u`, `p` and `N` dan memberikan jumlahan Riemann (Rieman sum) dengan menggunakan titik-titik tengah (*midpoints*) $x_k^*$ dari suatu partisi berukuran $N$:
    $$
    \int_0^u \frac{1}{\sqrt{1 + x^p}} dx \approx \sum_{k=1}^N \frac{1}{\sqrt{1 + (x_k^*)^p}} \Delta x
    $$
    Di sini $\Delta x = u/N$ dan $x_k^* = (x_k + x_{k-1})/2$ untuk titik-titik ujung $x_k = k \Delta x$.
